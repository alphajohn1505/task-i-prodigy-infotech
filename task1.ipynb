```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('indian_population.csv', encoding='latin1')
```

```python
class_distribution = df['Year'].value_counts()
```

```python
colors = ['skyblue', 'lightgreen', 'lightcoral', 'orange', 'lightblue', 'pink']
```

```python
plt.figure(figsize=(10, 6))
plt.bar(class_distribution.index, class_distribution.values, color=colors)
```

```python
plt.xlabel('Population')
plt.ylabel('Year')
plt.title('Distribution of Population over the years')
```

```python
for i, count in enumerate(class_distribution.values):
    plt.text(i, count + 5, str(count), ha='center', va='bottom', fontweight='bold')
```

```python
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```
